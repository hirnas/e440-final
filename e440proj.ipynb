```python
import intake
import xarray as xr
import intake_esm
#url = intake_esm.tutorial.get_url('google_cmip6')
#print(url)
url ="https://raw.githubusercontent.com/NCAR/intake-esm-datastore/master/catalogs/pangeo-cmip6.json"
```

```python
cat = intake.open_esm_datastore(url)
cat
```

```python
cat.df
```

```python
print(cat.unique()['activity_id'])
```

```python
ocat = cat_subset = cat.search(
    activity_id = 'OMIP'
)
ocat.df
```

```python
ocat_chl = ocat.search(
    variable_id = 'chl'
)
ocat_chl.df
dset_dict = ocat_chl.to_dataset_dict(
    xarray_open_kwargs={"consolidated": True, "decode_times": True, "use_cftime": True}
)
dset_dict
```

```python
[key for key in dset_dict.keys()][:10]
```

```python
dset_dict['OMIP.NOAA-GFDL.GFDL-CM4.omip1.Omon.gr'].chl.isel(time=1200, lev=0).plot()
```

```python
import dask

dask.config.set(scheduler='threads')
dask.config.set(num_workers=14)  # Adjust based on your CPU cores


#cesm_data_path = "http://thredds.atmos.albany.edu:8080/thredds/dodsC/CESMA/"
#atm_control = xr.open_dataset(cesm_data_path + "cpl_1850_f19/concatenated/cpl_1850_f19.cam.h0.nc")

#weight_factor = atm_control.gw / atm_control.gw.mean(dim='lat')

testplotdata = (dset_dict['OMIP.NOAA-GFDL.GFDL-CM4.omip1.Omon.gr'].chl.isel(lev=0)).mean(dim=('lat','lon')).groupby('time.year').mean()
testplotdata
```

```python
testplotdata.plot()
```
